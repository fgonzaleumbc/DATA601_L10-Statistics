# Introduction to Hypothesis Testing
[Return to Table of Contents](#Table-of-Contents)

Hypothesis testing allows us to **infer the probabilistic relationship between parameters and features in a dataset**. Essentially, it helps us test how plausible a hypothesis is given the observed data. It is an objective method for making decisions or drawing inferences from sample data (evidence).  

We typically use sample data to choose between two statements (hypotheses) about a population by comparing what we observe to what we would expect if the **Null Hypothesis** were true.

**Example Hypotheses:**
- Hypothesis 1: There is a difference between vaccinated and unvaccinated people => I accept getting vaccinated.  
- Hypothesis 2: There is no difference between vaccinated and unvaccinated people => I refuse to get vaccinated.

#### Hypothesis Testing Framework
There are always **two hypotheses**:  

**$H_0$: Null Hypothesis**  
- The starting assumption (what we assume to be true initially)  
- Typically states that there is **no difference, effect, or relationship**  

**$H_A$: Alternative (Research) Hypothesis**  
- What we aim to find evidence for  
- Typically states that there **is a difference, effect, or relationship**  

Example: Court Scenario
A jury must decide whether a person is guilty or innocent based on evidence.  
**Hypothesis Testing Framework:**  
- Null: The person is **innocent**  
- Alternative: The person is **not innocent** (i.e., guilty)  

> The null can only be rejected if there is enough evidence to doubt it. The jury can only convict if there is **beyond reasonable doubt** for the null of innocence. They do not know the personâ€™s true status and may make mistakes.

#### Types of Errors
(In machine learning, the confusion matrix is similar, showing true/false positives/negatives.)

- **Type I Error (False Positive):** Rejecting a null hypothesis that is actually true in the population.  
- **Type II Error (False Negative):** Failing to reject a null hypothesis that is actually false in the population.

# Statistical Tests (OPTIONAL)

This section provides a summary of various statistical tests used to infer relationships between different types of features.

### Chi-Square Test

The **Chi-squared test** is used to determine whether two categorical variables are related.  

The test statistic is calculated as the sum of the squared differences between observed (O) and expected (E) values:



The Chi-squared test assumes a **Chi-squared distribution**. Depending on the degrees of freedom, the test statistic can be converted into a **p-value**.



### P-Value

The **p-value** is the probability of obtaining a test statistic at least as extreme as the one observed, assuming the null hypothesis is true.  

- A p-value around 0.05 is commonly used as the threshold for statistical significance.  
- Lower values indicate stronger evidence against the null hypothesis ($H_0$) and support for the alternative hypothesis ($H_A$).

For the Chi-squared test, the p-value is computed using the Chi-squared distribution, which is skewed and depends on the degrees of freedom.

**Decision Rule:**
- If `P < 0.05`: reject $H_0$ â†’ evidence supports $H_A$ (there **is** an association)  
- If `P > 0.05`: do not reject $H_0$ â†’ no evidence of association

**Important Note on Expected Counts:**
- Check the number of cells with **expected counts < 5**.  
- If â‰¥ 20% of cells have expected counts < 5, the Chi-squared approximation may not be reliable.  
  - For **2x2 tables**, use **Fisherâ€™s Exact Test**.  
  - For larger tables (3x2, etc.), consider combining categories (if meaningful) to increase cell counts.

### ANOVA (Analysis of Variance)

**ANOVA** is used to compare the means of **more than two groups** (for two groups, a t-test is sufficient). It evaluates the variation **between groups** relative to the variation **within groups** using a variance-based **F-test**.  

Sometimes called an **omnibus test**, ANOVA tests a non-specific null hypothesis: that all group means are equal.


#### Key Points:
- **Types of ANOVA:**
  - **One-way ANOVA:** One factor (independent variable)  
  - **Two-way ANOVA:** Two factors (independent variables)
- **Univariate ANOVA:** Only one dependent variable.  
  - For multiple dependent variables: **MANOVA**  
  - For additional continuous covariates: **ANCOVA**  
- **Repeated Measures ANOVA:** Used when measurements are taken multiple times on the same subjects.


#### Variance Estimation in ANOVA

- **Between-group variation:** Mean Square due to Treatments (MST or MS_between)  
- **Within-group variation:** Mean Square due to Error (MSE or MS_within)

**Within-group variation:**
- Residual = difference between an individual observation and its group mean  
- SS_within = sum of squared residuals  

**Between-group variation:** Differences between each group mean and the overall mean.

#### Hypotheses

- **Null hypothesis ($H_0$):** All group means are equal  
$H_0$: $\mu_1 = \mu_2 = \cdots \mu_p$

- **Alternative hypothesis ($H_A$):** At least one group mean is different  
$H_A$: At least, one group mean is different from other groups (i.e., All $\mu$'s are not equal). <br>

> The omnibus F-test evaluates $H_0$ across all groups. If rejected, **post-hoc tests** are used to identify which groups differ.

#### Assumptions for ANOVA
1. **Normality:** Residuals are approximately normally distributed (check with Shapiro-Wilk test or histogram)  
2. **Homoscedasticity (Equal variances):** Variances are similar across groups (Leveneâ€™s, Bartlettâ€™s, or Brown-Forsythe test)  
3. **Independence:** Observations are independent of each other  
4. **Continuous Dependent Variable:** Ordinal/rank variables may violate normality; consider non-parametric tests like Mann-Whitney U or Kruskal-Wallis if assumptions fail  

> ANOVA is powerful when assumptions hold. Violation of normality with equal variances slightly reduces power but has little effect on Type I error.


### Regression Analysis

**Regression** (e.g., linear, multiple linear, logistic, etc.) is used when analyzing relationships between **one or more independent and dependent variables**.  

Regression is particularly useful to:
- Identify significant relationships between variables  
- Predict the value of a dependent variable for given values of one or more independent variables  

The method involves estimating a **line of best fit** that minimizes the **sum of squared residuals** (differences between observed and predicted values).

#### Simple Linear Regression

- Examines the relationship between **two continuous (scale) variables**  
- Produces a straight-line equation of the form:  

$$
y = \beta_0 + \beta_1 x
$$

- The **coefficient of determination ($R^2$)** indicates how well the model explains the variability of the dependent variable and how reliable the predictions are.


# Test Selection (IMPORTANT)
[Return to Table of Contents](#Table-of-Contents)

This section covers **how to select the appropriate statistical test** based on the types of variables in your data and the relationships you want to evaluate.

### Variables

**Independent variables** (explanatory/predictor) influence a **dependent variable** (outcome).  

**Types of variables:**

- **Scale (numeric):** Continuous or discrete numeric values  
- **Categorical:** Variables with distinct categories, which can be:  
  - **Ordinal:** Categories with a natural order (e.g., rating scales)  
  - **Nominal:** Categories without any natural order (e.g., colors, gender)

### Types of Statistical Tests

Statistical tests are generally classified as:

- **Parametric Tests:** are statistical tests that make specific assumptions about the **underlying distribution** of the data, usually assuming it follows a **normal distribution**. Parametric tests estimate parameters (like the mean and standard deviation) of the population from the sample data, hence the name â€œparametric.â€ Key characteristics of parametric tests:
    - Require the dependent variable to be **continuous (scale)**  
    - Assume **homogeneity of variances** across groups  
    - Often more **powerful** than non-parametric tests if assumptions are met  
    - Assume the data follows a specific distribution (e.g., normal distribution)  
    - Examples: t-test, ANOVA, linear regression  

- **Non-Parametric Tests:** are statistical tests that **do not assume a specific distribution** for the data. They are useful when the assumptions of parametric tests (e.g., normality or equal variances) are violated. Non-parametric tests are sometimes called **distribution-free tests** because they do not rely on population parameters like the mean or standard deviation. Key characteristics of non-parametric tests:
    - Can be used with **ordinal, nominal, or skewed scale data**  
    - Often based on **ranks or signs** rather than the raw values  
    - Less sensitive to outliers and non-normal distributions  
    - Do not assume a specific distribution  
    - Often based on **ranks** or **signs** rather than raw data  
    - Examples: Mann-Whitney U test, Kruskal-Wallis test, Wilcoxon signed-rank test, Spearman correlation

# Correlation and Covariance
[Return to Table of Contents](#Table-of-Contents)

<b>Correlation:</b> A measure of the strength of a relationship between data. Measures both the strength and direction of the relationship between two variables. It tells us how changes in one variable are associated with changes in another.

$$\mathrm{r}(X, Y) = \frac{\sum\limits_{i}^{N}{(X_i â€“ \mu_x)(Y_i â€“ \mu_y)}}{{\sqrt{\sum\limits_{i}^{N}{(X_i â€“ \mu_x)^2}\sum\limits_{i}^{N}{(Y_i â€“ \mu_y)^2}}}} $$

<b>Covariance:</b> A measure of the variance between two variables. Measures how two variables change together. Indicates the direction of the linear relationship between variables.  It indicates whether increases in one variable are associated with increases or decreases in the other.

$$\mathrm{cov}(X, Y) = \frac{1}{{N}} \sum\limits_{i}^{N}{(X_i â€“ \mu_x)(Y_i â€“ \mu_y)}$$

Both measure the relationship and the dependency between two variables. Typically, shown in a heatmap form with the values of correlation being standardized while the covariance IS NOT standardized. Note that correlation does not equal causation.

```python
# Let's plot a few features: "Passenger class" and "Survived or not".
plt.figure(figsize=(10,2.5))

plt.subplot(121)
sns.countplot(data = df, x='Pclass')
plt.title("Passenger Class")

plt.subplot(122)
sns.countplot(data = df, x='Survived')
plt.title("Deceased (0) or Survived (1)")

plt.show()
```

```python
# Let's do a pivot table of those same features: "PClass" and "Survived"
PClass_survd = pd.pivot_table(df,index=['Pclass'],columns=['Survived'],aggfunc='size')
PClass_survd
```

38.38% of total passeneger survived

### Analysis of Class
[Return to Table of Contents](#Table-of-Contents)

Null hypothesis: The survival does not depend on the class the passengers were travelling.

If this is true then we expect the distributions of survaivability to have the same proportions in all columns regardless of the class. Where the expected value for survability is 38.38% as previously calculated.

```python
# To calculated the expected value.
display(pct_class.to_frame()) # Percent per class
display(pct_survived.to_frame().T) # Percent survivability
```

```python
# Total percent Expected survivability by class. 
# Note the @ is called a decorator. Is another way to reference a variable or a function.
pct_class.to_frame()@(pct_survived.to_frame().T)#*891 
# Mutliplying by 891 converts back to expected count survavibility. See next cell.
```

```python
# This is the number of people in all classes and expected survival history.
# This converts the both the pct_class and pct_survived to a dataframe and combines the values. 
expected_survived = round(pct_class.to_frame()@(pct_survived.to_frame().T)*891)
expected_survived
# This is the expected ratios assuming taht survival distribution is independent of class.
```

```python
# Getting the tables of Class/Survival Observed/Expected as heatmaps.

plt.figure(figsize=(10,4))
sns.heatmap(PClass_survd,annot=True, fmt='g',square=True,cmap='hot')
plt.title('Observed',fontsize=16)
plt.show();

plt.figure(figsize=(10,4))
sns.heatmap(expected_survived, annot=True, fmt='g',square=True,cmap='hot')
plt.title('Expected',fontsize=16)
plt.tight_layout()

plt.show();
# Note the values in class 1. The ratio for survived is inverted in the observed vs. expected.
```

```python
# Let's calculate the chi-square test
chi2_stat, p_val, dof, ex = stats.chi2_contingency(PClass_survd)

print("Chi square value is ",chi2_stat)
print("P value is",p_val)
print("Degrees of Freedom:",dof)
```

P value is almost 0! we can reject the null hypothesis with confidence!

### Example-1
[Return to Table of Contents](#Table-of-Contents)

Compare two unrelated samples. Data was collected on the weight loss of 16 women and 20 men enrolled in a weight reduction program.
At $\alpha$ = 0.05, test whether the weight loss of these two samples is different.

```python
Weight_loss_Male   = [3.69, 4.12, 4.65, 3.19,  4.34, 3.68, 4.12, 4.50, 3.70, 3.09,3.65, 
                      4.73, 3.93, 3.46, 3.28, 4.43, 4.13, 3.62, 3.71, 2.92]
Weight_loss_Female = [2.99, 1.80, 3.79, 4.12, 1.76, 3.50, 3.61, 2.32, 3.67, 4.26, 4.57, 
                      3.01, 3.82, 4.33, 3.40, 3.86]
```

#### Step 1: Define null and alternative hypotheses
In testing whether weight reduction of female and male are same,the null hypothesis states that mean weight reduction, $\mu{M}$ equals $\mu{F}$. The alternative hypthesis states that the weight reduction is different for Male and Female, $\mu{M}$ $\neq$ $\mu{F}$

* $H_0$: $\mu{M}$ - $\mu{F}$ =      0
* $H_A$: $\mu{M}$ - $\mu{F}$ $\neq$  0

#### Step 2: Decide the significance level
Here we select $\alpha$ = 0.05 and sample size < 30 and population standard deviation is not known.

#### Step 3: Identify the test statistic

* We have two samples and we do not know the population standard deviation.
* Sample sizes for both samples are not same.
* The sample is not a large sample, n < 30. So you use the t distribution and the `t-STAT` test statistic for two sample unpaired test.

#### Step 4: Calculate the p - value and test statistic

** We use the scipy.stats.ttest_ind to calculate the t-test for the means of TWO INDEPENDENT samples of scores given the two sample observations. This function returns t statistic and two-tailed p value.**

** This is a two-sided test for the null hypothesis that 2 independent samples have identical average (expected) values. This test assumes that the populations have identical variances.**

```python
t_statistic, p_value  =  stats.ttest_ind(Weight_loss_Male, Weight_loss_Female)
print('P Value %1.3f' % p_value)
```

#### Step 5:  Decide to reject or accept null hypothesis

In this example, p value is 0.076 and it is more than 5% level of significance

The statistical decision is to accept the null hypothesis at 5% level of significance.

There is no sufficient evidence to reject the null hypothesis that the weight loss of these men and women is same.

### Study-1: Poor Marriages
[Return to Table of Contents](#Table-of-Contents)

Let's compare the relationship of a poor marriage and affairs.

#### Chi-Square 
Chi-square test of independence: The hypothesis-testing procedure appropriate when both the independent variable and the dependent variables are categorical.

#### Does having a poor marriage lead to affairs?

Let's create categorical data and add them as two new columns into our dataframe. <br>
 - Poor Marriage = yes, if rate_marriage <= 2 <br>
 - Affair = yes, if affair > 0

```python
# Creat two derived features with poor_marriage_yn and affairs_yn.
df_fair_v2 = df_fair.assign(poor_marriage_yn = (df_fair.rate_marriage <= 2), 
                            affairs_yn = (df_fair.affairs > 0))
df_fair_v2.head()
```

```python
df_fair_v2.poor_marriage_yn.value_counts()
```

```python
df_fair_v2.affairs_yn.value_counts()
```

```python
df = (df_fair_v2.groupby(['poor_marriage_yn', 'affairs_yn'])[['affairs']].count().unstack().droplevel(axis=1, level=0))
df
```

Note that we know the numbers of poor marriages, we know the number of affairs but we don't know whether those pieces of data come from the same couples. In other words, these are two independent groups of observations.

Hypothesis: Assume the affair proportions are equal among poor marriages and non-poor marriages. <br>
In other other our words: <br>
Null hypothesis is that "there is no correlation between marriage poornees and having affairs.

```python
chi2, p, dof, ex = stats.chi2_contingency(df, correction=False)
print('chi2:', chi2)
print('p-value:', p)
```

p-value is 0 (in fact, it is a very small number, almost zero). This means that we have to reject the null hypothesis!<br>

In fact, we could come up with the same claim just by looking at the normalized/scaled version of df. See below. Affair ratio is 30% among "non-poor marriages" and 66% among "poor marriages". So they don't have same means!

```python
df_pct = df.apply(axis=1, func=lambda r: r/r.sum()) #
df_pct
```

Let's visualize this with SNS

```python
sns.countplot(data=df_fair_v2,x='poor_marriage_yn', hue='affairs_yn', saturation=0.9);
```

#### Welch's t-test
We could have used Welch's t-test to answer the same question.

In this case, we will compare the average affair count among two groups: poor marriages and non-poor marriages!

```python
# First, let's check the describe function.
# This is the statistics for affairs column grouped by poor_marriege_yn.
df_fair_v2.groupby('poor_marriage_yn').affairs.describe().T
```

```python
# They definitely have very different means!
# Affair values among poor marriages
a = df_fair_v2[df_fair_v2.poor_marriage_yn].affairs
# Affair values among non-poor marriages
b = df_fair_v2[~df_fair_v2.poor_marriage_yn].affairs

# Welch's T-Test
tstats, pvalue = stats.ttest_ind(a, b, equal_var=False) # Welch's T-Test when "equal_var = False". See documentation.
format(pvalue, '.3g') # P-value
```

P-value is very small number, almost zero. This means that we have to reject the null hypothesis! We can visualize the means and as previously confirm they are not the same as previously thought.

```python
# Let's visualize the data and we can observe that the means are 
# not the same as previously thought.
sns.pointplot(data = df_fair_v2, x='poor_marriage_yn', y='affairs');
```

### Chi-Square 

Null Hypothesis: There is no difference among farmers and business people in terms of affair averages.

```python
df_v3 = df_fair_v2[df_fair_v2.occupation.isin([2, 6])] # Occupation 2 and 6, farmers and business occupations respectively.
df_v3.head()
```

### Fisher's exact test
[Return to Table of Contents](#Table-of-Contents)

The null hypothesis is that the true odds ratio of the populations underlying the observations is one, and the observations were sampled from these populations under a condition: the marginals of the resulting table must equal those of the observed table. The statistic returned is the unconditional maximum likelihood estimate of the odds ratio, and the p-value is the probability under the null hypothesis of obtaining a table at least as extreme as the one that was actually observed. There are other possible choices of statistic and two-sided p-value definition associated with Fisherâ€™s exact test; please see the Notes for more information.

The Fisher EXact function returns the prior odds ratio (not a posterior estimate) and probability under the null hypothesis of obtaining a table at least as extreme as the one that was actually observed.

# Welch's t-test vs. Student's t-test Example (OPTIONAL)
[Return to Table of Contents](#Table-of-Contents)

Student's t-test assumes the two populations have the same variance! If you know this is true, then you can do the student's t-test (i.e., `stats.ttest_ind(a, b)`)

However, most of the time, this is not true and we have to do the Welch's t-test (i.e. `stats.ttest_ind(a, b, equal_var=False)`).

```python
# stats.norm.rvs creates normal continuous random variables.
# The location (loc) keyword specifies the mean.
# The scale (scale) keyword specifies the standard deviation.

# Let's create two sample sets with the same mean and scale but different sizes
group_ctl = stats.norm.rvs(loc=120, scale=5, size=1000)
group_exp = stats.norm.rvs(loc=120, scale=5, size=20)
```

```python
name_pvalue_pairs = [
    ("Student's t-test", stats.ttest_ind(group_ctl, group_exp).pvalue),
    ("Welch's t-test", stats.ttest_ind(group_ctl, group_exp, equal_var=False).pvalue)]
name_pvalue_pairs.sort(key=lambda x: x[1])

for name, p_value in name_pvalue_pairs:
    print(f'{name:26} {p_value:.4f}  {p_value < 0.05}')
```

# ANOVA Example (OPTIONAL)
[Return to Table of Contents](#Table-of-Contents)

ANOVA works using the following steps:
- Check sample sizes: equal number of observation in each group
- Calculate Mean Square for each group (MS) (SS of group/level-1); level-1 is a degrees of freedom (df) for a group
- Calculate Mean Square error (MSE) (SS error/df of residuals)
- Calculate F-value (MS of group/MSE)
- Calculate p-value based on F value and degrees of freedom (df) 

One way ANOVA Documentation:
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.f_oneway.html

```python
# Data values for each treatment A, B, C, D.
anova_array = [[25,45,30,54],
               [30,55,29,60],
               [28,29,33,51],
               [36,56,37,62],
               [29,40,27,73]]
```

```python
df = pd.DataFrame(anova_array, columns =['A', 'B', 'C','D'])
df.head()
```

```python
# Reshape the dataframe to make it suitable for statsmodels package 
df_melt = pd.melt(df.reset_index(), id_vars=['index'], value_vars=['A', 'B', 'C', 'D'])
# Replace column names
df_melt.columns = ['index', 'treatments', 'value']
print(df_melt.shape)
df_melt.head()
```

```python
# Boxplot for each treatment categorical value.
sns.boxplot(x='treatments', y='value', data=df_melt, color='#99c2a2')
sns.swarmplot(x="treatments", y="value", data=df_melt, color='#7d0013')
plt.show()
```

```python
# stats f_oneway functions takes the groups as input and returns ANOVA F and p value
fvalue, pvalue = stats.f_oneway(df['A'], df['B'], df['C'], df['D'])
print(fvalue, pvalue)
```

The p-value obtained from ANOVA analysis is significant (p < 0.05), and therefore, we conclude that there are significant differences among treatments.

Note on F-value: F value is inversely related to p-value. Higher F-value (greater than F critical value) indicates a significant p-value.

```python
# Ordinary Least Squares (OLS) model
model = ols('value ~ C(treatments)', data=df_melt).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

From ANOVA analysis, we know that treatment differences are statistically significant, but ANOVA does not tell which treatments are significantly different from each other. To know the pairs of significant different treatments, we can perform multiple pairwise comparison (post hoc comparison) analysis for all unplanned comparison using Tukeyâ€™s honestly significantly differenced (HSD) test.
